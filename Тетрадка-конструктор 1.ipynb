## Cоздание новых переменных

### Создаем переменную, у которой значения основаны на значениях другой переменной

```python
# Выводим уникальные значения исходной переменной, в данном случае - значения переменной job_position
print(data3['job_position'].unique())
```

```python
# Затем создаем словарь, в котором ключом будет значение исходной переменной job_position, 
# а значением - значение будущей переменной avrzarplata
dct = {'UMN': 51000, 'SPC': 63000, 'INP': 55000, 'DIR': 60000, 
       'ATP': 46000, 'PNA': 71000, 'BIS': 86000, 'WOI': 76000,
       'NOR': 54000,'WRK': 77000, 'WRP': 75000, 'PNV': 67000, 
       'BIU': 43000, 'PNI': 69000, 'HSK': 74000, 'PNS': 44000, 
       'INV': 88000, 'ONB': 62000}
# Создаем новую переменную avrzarplata, у которой значения сопоставлены значениям переменной job_position
data3['avrzarplata'] = data3['job_position'].map(dct)
data3.head()
```

### Создаем переменную, которая является отношением двух переменных

```python
# Создаем переменную, которая является отношением суммы кредита к ежемесячному заработку
data3['ratio'] = data3['credit_sum'] / data3['monthly_income']
data3.head()
```

### Создаем переменную, у которой каждое значение - частота наблюдений в категории переменной

```python
# Создаем переменную region_cnt, у которой каждое значение - частота наблюдений в категории переменной living_region
data3['region_cnt'] = data3['living_region'].map(data3['living_region'].value_counts())
data3.head()
```

### Создаем переменную, у которой каждое значение - среднее значение количественной переменной, взятое по уровню категориальной переменной

```python
# Пишем функцию, создающую переменную, у которой каждое значение будет средним значением количественной
# переменной (real_feature), взятым по уровню категориальной переменной (cat_feature)
def code_mean(data, cat_feature, real_feature):
    return (data[cat_feature].map(data.groupby(cat_feature)[real_feature].mean()))

# Создаем переменную, у которой каждое значение - среднее значение monthly_income 
# в категории переменной living_region
data3['region_mean_income'] = code_mean(data3, 'living_region', 'monthly_income')
data3.head()
```

### Создаем категориальную переменную в результате конъюнкции двух переменных

```python
# Пишем функцию, которая создает переменную в результате конъюнкции
# переменных feature1 и feature2
def make_conj(data, feature1, feature2):
    data[feature1 + ' + ' + feature2] = data[feature1].astype(str) + ' + ' + data[feature2].astype(str)
    return (data)
make_conj(data3, 'education', 'marital_status')
data3.head()
```

### Создаем переменную, у которой каждое значение - логарифм значения количественной переменной

```python
# Создаем переменную log_income, у которой каждое значение - логарифм значения
# переменной monthly_income
data3['log_income'] = np.log(data3['monthly_income'])
data3.head()
```

### Создаем количественную переменную, у которой каждое значение - усредненное значение нескольких количественных переменных

```python
# Создаем переменную mean_age_tenure, у которой каждое значение - усредненное значение 
# переменных age и credit_month 
data3['mean_age_tenure'] = data3[['age', 'credit_month']].mean(axis=1)
data3.head()
```

### Выполняем категоризацию (биннинг) количественной переменной на основе интервалов, заданных вручную

```python
# Вычисляем минимальное значение переменной monthly_income
min_income = data3['monthly_income'].min()
min_income
```

```python
# Вычисляем максимальное значение переменной monthly_income
max_income = data3['monthly_income'].max()
max_income
```

```python
# Задаем точки, в которых будут находится границы категорий 
# (от 5000 до 50000, от 50000 до 200000, от 200000 до 950000) 
bins = [-np.inf, 50000, 200000, np.inf]
# Задаем метки для категорий будущей переменной
group_names = ['Low', 'Average', 'High']
# Осуществляем биннинг переменной monthly_income 
# и записываем результаты в новую переменную incomecat
data3['incomecat'] = pd.cut(data3['monthly_income'], bins, labels=group_names)
data3.head()
```

```python
# Пишем функцию, которая категоризирует
# переменную monthly_income
def income_category(monthly_income):
    '''
    < 20000 -> 1
    >= 20000, <40000 -> 2
    >= 40000 -> 3
    '''
    if monthly_income < 20000:
        return 1
    elif monthly_income < 40000:
        return 2
    else:
        return 3
```

```python
# создаем новую переменную income_category
income_categories = [income_category(monthly_income) for monthly_income in data3.monthly_income]
data3['income_category'] = income_categories
data3.head()
```

```python
# а еще можно воспользоваться методом .apply()
data3['income_category2'] = data3['monthly_income'].apply(income_category)
data3.head()
```

### Выполняем категоризацию (биннинг) количественной переменной на основе децилей

```python
# Осуществляем биннинг переменной monthly_income на основе децилей и записываем результаты 
# в новую переменную income_decile, не создавая меток категорий
data3['income_decile'] = pd.qcut(data3['monthly_income'], 10, labels=False)
data3.head()
```

### Создаем бинарную переменную на основе значений количественной переменной

```python
# создаем новую переменную retired, которая принимает значение "Yes",
# если значение переменной age больше 60, и значение "No" в противном случае
data3['retired'] = np.where(data3['age']>=60, 'Yes', 'No')
data3.head()
```

### Создаем бинарную переменную на основе значений количественных переменных

```python
# создаем новую переменную age_inc, которая принимает значение "Yes",
# если речь идет о клиентах старше 35 лет и с суммой кредита свыше 10000, 
# и значение "No" в противном случае
data3['age_inc'] = np.where((data3['age'] > 35) & (data3['credit_sum'] > 10000), 'Yes', 'No')
data3.head(10)
```

```python
# создаем новую переменную age_inc, которая принимает значение "Yes",
# если речь идет о клиентах старше 35 лет или с суммой кредита свыше 13000, 
# и значение "No" в противном случае
data3['age_inc2'] = np.where((data3['age'] > 35) | (data3['credit_sum'] > 13000), 'Yes', 'No')
data3.head(10)
```

```python
# Создаем список из 3 цветов по количеству будущих секторов
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B"]

# Создаем круговую диаграмму
plt.pie(
    # используем частоты значений переменной incomecat
    counts,
    # задаем список меток
    labels=["Mar", "Unm", "Div", "Civ", "Wid"],
    # без использования тени
    shadow=False,
    # задаем список цветов
    colors=colors,
    # один сектор отделяем от других
    explode=(0, 0, 0, 0, 0.20),
    # задаем начальный угол в градусах
    startangle=90,
    # выводим значения в виде процентов
    autopct='%1.1f%%'
    )

plt.axis('equal')
plt.tight_layout()
plt.show()
```
